# Transformation operation

### airline

```python
air_transformer = Pipeline(steps=[
	("imputer", SimpleImputer(strategy="most_frequent")),
	("grouper", RareLabelEncoder(tol=0.1, replace_with="Other", n_categories=2)),
	("encoder", OneHotEncoder(sparse_output=False, handle_unknown="ignore"))
])

air_transformer.fit_transform(X_train.loc[:, ["airline"]])#.airline.value_counts()
```

### date_of_journey

```python
feature_to_extract=["month","week","day_of_week","day_of_year"]
doj_transform=Pipeline(steps=[
    ("dt",DatetimeFeatures(features_to_extract=feature_to_extract,yearfirst=True,format="mixed")),("scalar",MinMaxScaler())
])
doj_transform.fit_transform(X_train.loc[:,["date_of_journey"]])
```

### source & destination

```python
loc=X_train.loc[:,["source","destination"]]
location_pipe1=Pipeline(steps=[("grouper",RareLabelEncoder(tol=0.1,replace_with="Other",n_categories=2)),
                               ("encoder",MeanEncoder()),
                               ("transform",PowerTransformer())])
location_pipe1.fit_transform(loc,y_train)
```

```python
np.union1d(
    X_train.source.unique(),
    X_train.destination.unique()
)
```

```python
def is_north(X):
    columns=X.columns.to_list()
    north_cities=["Delhi","Kolkata","Mumbai","New Delhi"]
    return (
        X.assign(**{
            f"{col}_is_north": X.loc[:,col].isin(north_cities).astype(int)
            for col in columns
        })
        .drop(columns=columns)
    )
FunctionTransformer(func=is_north).fit_transform(loc)
```

```python
location_trans=FeatureUnion(transformer_list=[
    ("part1",location_pipe1),
    ("part2",FunctionTransformer(func=is_north))
])
location_trans.fit_transform(loc,y_train)
```

### dep_time & arrival_time

```python
time_subset=X_train.loc[:,["dep_time","arrival_time"]]
time_pipe1=Pipeline(steps=[
    ("dt",DatetimeFeatures(features_to_extract=["hour","minute"])),
    ("scale",MinMaxScaler())
])
time_pipe1.fit_transform(time_subset)
```

```python
def part_of_day(X,morning=4,noon=12,eve=16,night=20):
    columns=X.columns.to_list()
    X_temp=X.assign(**{
        col:pd.to_datetime(X.loc[:,col]).dt.hour
        for col in columns
    })
    return (
        X_temp
        .assign(**{
            f"{col}_part_of_day":np.select([X_temp.loc[:,col].between(morning,noon,inclusive="left"),
                                           X_temp.loc[:,col].between(noon,eve,inclusive="left"),
                                           X_temp.loc[:,col].between(eve,noon,inclusive="left")],
                                           ["morning","afternoon","evening"],
                                           default="night")
            for col in columns
        })
        .drop(columns=columns)
    )
FunctionTransformer(func=part_of_day).fit_transform(time_subset)
```

```python
time_pipe2=Pipeline(steps=[("part1",FunctionTransformer(func=part_of_day)),
                           ("part2",CountFrequencyEncoder()),
                           ("scalar",MinMaxScaler())])
time_pipe2
```

```python
time_trans=FeatureUnion(transformer_list=
                        [("part1",time_pipe1),
                        ("part2",time_pipe2)])
time_trans.fit_transform(time_subset)
```

### duration

```python
X_train.duration
```

```python
class RBFPercentileSimilarity(BaseEstimator, TransformerMixin):
    def __init__(self, variables=None, percentiles=[0.25, 0.5, 0.75], gamma=0.1):
            self.variables = variables
            self.percentiles = percentiles
            self.gamma = gamma


    def fit(self, X, y=None):
        if not self.variables:
            self.variables = X.select_dtypes(include="number").columns.to_list()
        
        self.reference_values_ = {
			col: (
				X
				.loc[:, col]
				.quantile(self.percentiles)
				.values
				.reshape(-1, 1)
			)
			for col in self.variables
		}
        return self
    
    def transform(self, X):
     objects = []
     for col in self.variables:
         columns = [f"{col}_rbf_{int(percentile * 100)}" for percentile in self.percentiles]
         obj = pd.DataFrame(
				data=rbf_kernel(X.loc[:, [col]], Y=self.reference_values_[col], gamma=self.gamma),
				columns=columns
			)
         objects.append(obj)
         return pd.concat(objects, axis=1)
		
```

```python
X_duration = X_train[["duration"]]
RBFPercentileSimilarity().fit_transform(X_duration)
```

```python
def duration_category(X, short=180, med=400):
	return (
		X
		.assign(duration_cat=np.select([X.duration.lt(short),
									    X.duration.between(short, med, inclusive="left")],
									   ["short", "medium"],
									   default="long"))
		.drop(columns="duration")
	)
```

```python
def is_over(X, value=1000):
	return (
		X
		.assign(**{
			f"duration_over_{value}": X.duration.ge(value).astype(int)
		})
		.drop(columns="duration")
	)
```

```python
duration_pipe1 = Pipeline(steps=[
	("rbf", RBFPercentileSimilarity()),
	("scaler", PowerTransformer())
])

duration_pipe2 = Pipeline(steps=[
	("cat", FunctionTransformer(func=duration_category)),
	("encoder", OrdinalEncoder(categories=[["short", "medium", "long"]]))
])

duration_union = FeatureUnion(transformer_list=[
	("part1", duration_pipe1),
	("part2", duration_pipe2),
	("part3", FunctionTransformer(func=is_over)),
	("part4", StandardScaler())
])

duration_transformer = Pipeline(steps=[
	("outliers", Winsorizer(capping_method="iqr", fold=1.5)),
	("imputer", SimpleImputer(strategy="median")),
	("union", duration_union)
])

duration_transformer.fit_transform(X_train.loc[:, ["duration"]])
```

### total_stops

```python
def is_direct(X):
	return X.assign(is_direct_flight=X.total_stops.eq(0).astype(int))


total_stops_transformer = Pipeline(steps=[
	("imputer", SimpleImputer(strategy="most_frequent")),
	("", FunctionTransformer(func=is_direct))
])

total_stops_transformer.fit_transform(X_train.loc[:, ["total_stops"]])
```

### additional_info

```python
info_pipe1 = Pipeline(steps=[
	("group", RareLabelEncoder(tol=0.1, n_categories=2, replace_with="Other")),
	("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

info_pipe1.fit_transform(X_train.loc[:, ["additional_info"]])
```

```python
def have_info(X):
	return X.assign(additional_info=X.additional_info.ne("No Info").astype(int))
```

```python
info_union = FeatureUnion(transformer_list=[
	("part1", info_pipe1),
	("part2", FunctionTransformer(func=have_info))
])
```

```python
info_transformer = Pipeline(steps=[
	("imputer", SimpleImputer(strategy="constant", fill_value="unknown")),
	("union", info_union)
])

info_transformer.fit_transform(X_train.loc[:, ["additional_info"]])
```

# column transformer

```python
column_tranformer=ColumnTransformer(transformers=[("air",air_transformer,["airline"]),
                                                  ("doj",doj_transform,["date_of_journey"]),
                                                  ("location",location_trans,["source","destination"]),
                                                  ("time",time_trans,["dep_time","arrival_time"]),
                                                  ("dur", duration_transformer, ["duration"]),
                                                  ("stops", total_stops_transformer, ["total_stops"]),
                                                  ("info", info_transformer, ["additional_info"])
                                                  ],remainder="passthrough")
column_tranformer.fit_transform(X_train,y_train)
```

# Feature selection

```python
estimator = RandomForestRegressor(n_estimators=10, max_depth=3, random_state=42)

selector = SelectBySingleFeaturePerformance(
	estimator=estimator,
	scoring="r2",
	threshold=0.1
) 
```

# Putting it all Together

```python
preprocessor = Pipeline(steps=[
	("ct", column_tranformer),
	("selector", selector)
])

preprocessor.fit(X_train, y_train)
preprocessor.transform(X_train)
```

# Visualizations

```python
feature_performances = preprocessor.named_steps["selector"].feature_performance_
feature_performances
```

```python
sorted_feat_imp = dict(sorted(feature_performances.items(), key=lambda val: val[1]))
sorted_feat_imp
```

```python
THRESHOLD = 0.1

selected_bar = None
dropped_bar = None
colors = ["red" if score < THRESHOLD else "green" for score in sorted_feat_imp.values()]


fig, ax = plt.subplots(figsize=(15, 4)) 

for i, (feature, score) in enumerate(sorted_feat_imp.items()):
	params = dict(
		x=i,
		height=score,
		edgecolor="black",
		alpha=0.5
	)
	
	if score < THRESHOLD:
		bar = ax.bar(
			color="red",
			**params
		)
		if not dropped_bar:
			dropped_bar = bar[0]
	else:
		bar = ax.bar(
			color="green",
			**params
		)
		if not selected_bar:
			selected_bar = bar[0]

thresh_line = ax.axhline(
	y=0.1,
	color="black",
	linestyle="--"
)

ax.set_xticks(
	ticks=range(len(sorted_feat_imp)),
	labels=list(sorted_feat_imp.keys()),
	rotation=30,
	ha="right"
)

ax.set(
	xlabel="Feature",
	ylabel="Score",
	title="Feature Selection Scores"
)

ax.legend(
	handles=[selected_bar, dropped_bar, thresh_line],
	labels=["Selected", "Dropped", "Threshold"],
	loc="upper left"
)

plt.show()
```

```python
import os
import pandas as pd

DATA_PREFIX = r'C:\Users\as\OneDrive\Desktop\projects\Flight Price prediction\data'
train=pd.read_csv(os.path.join(DATA_PREFIX,"train.csv"))
val=pd.read_csv(os.path.join(DATA_PREFIX,"val.csv"))
test=pd.read_csv(os.path.join(DATA_PREFIX,"test.csv"))


def get_file_name(name):
    return os.path.join(DATA_PREFIX, f"{name}-pre.csv")

def save_transformed_data(data, name, preprocessor):

    X_transformed = preprocessor.transform(data.drop(columns="price"))
    y = data.price.copy()
    
    final_data = pd.DataFrame(X_transformed)
    final_data['price'] = y
    
    file_name = get_file_name(name)
    final_data.to_csv(file_name, index=False, header=True)
    print(f"Data exported to {file_name}")
    
preprocessor.fit(train.drop(columns="price"), train.price)

save_transformed_data(train, "train", preprocessor)
save_transformed_data(val, "val", preprocessor)
save_transformed_data(test, "test", preprocessor)
```
